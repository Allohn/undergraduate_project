# Part 1: Drunken Sailor - 2D Random Walk

```python
def grid_create(steps, D): #Function to create the movement grid
    dim = (2 * steps) + 1 #twice number of steps to account for movement only to the right,left,up,down.+1 for W in centre
    
    if D == 2: #Option for 3D or 2D
        grid = np.zeros((dim, dim))
    else:
        grid = np.zeros((dim,dim,dim))
    
    return grid

def distance_travelled(steps): #Function to run 1 random walk with N steps and calculate distance
    grid = grid_create(steps, 2)
    
    grid[steps, steps] = 1 #Places the walker (W) in the centre of the grid.
    i_index = np.array([steps, steps]) #Keeps track of W position
    steps_taken = 0
    
    index = np.array([steps, steps])
    while steps_taken < steps:
        direction = rng.randint(0,100)#Generates an integer between 0 and 100
        
        if direction in range(25):#If [0-25) move north by 1
            index += np.array([0, 1])
            
        elif direction in range(25,50):#If [25-50) move east by 1
            index += np.array([1, 0])
            
        elif direction in range(50,75):
            index += np.array([0, -1])
            
        else:
            index += np.array([-1, 0])
            
        #print(index)
        #grid[index[0], index[1]] += 1
            
        steps_taken += 1
        
        #print(grid)  
        #print("")
        
        #Commented lines used to visualise the walk
       
    #print(grid)
    #Now calculate the distance travelled.
    displacement = index - i_index  
    distance = np.linalg.norm(displacement)
    
    return distance

def mean_distance_travelled(repetitions, steps): #Allows us to get a mean distance travelled for N steps
    distance_list = []
    for i in range(repetitions):
        distance_list.append(distance_travelled(steps))
        
    return np.mean(distance_list), np.std(distance_list) #Averages distance travelled when N steps are taken over a number of repetitions.

def distance_distribution(maxsteps, repetition):
    distribution = [] #Store the mean distance travelled in N steps where N is the index in the list + 1
    errors = []
    
    for i in tqdm(range(1, maxsteps + 1)):
        distribution.append(mean_distance_travelled(repetition, i)[0])
        errors.append(mean_distance_travelled(repetition, i)[1])
        time.sleep(0.1)
        
    return distribution
```

As this data seems to follow a power law, let's try to determine the form of it taking that the general form of a power law relation is $$ D = kN^{n}$$

When the log of both sides is taken, we find that there is a linear relation such that $$\log{D} = \log{k} + n\log{N}$$
Thus, plotting a linear regression model over this will allow us to determine n as the slope of such a regression and the value of k as the exponential of the intersect.

```python
def linear_regression(logN, n, logk):
    return logk + n * logN

logN = np.log(Nvalues)
params = curve_fit(linear_regression, logN, np.log(data2d))

print("n = " + f"{params[0][0]:.3f}")
print("k = " + f"{np.exp(params[0][1]):.3f}")
```

```python
linearfit = np.vectorize(linear_regression)

plt.plot(logN, linearfit(logN, params[0][0], params[0][1]) , color="red", label="Linear Regression")
plt.plot(logN, np.log(data2d), color="lightblue", label="Simulation Data")
plt.title("Linear Regression overlaid on a log-log plot of the data", fontsize = 15)
plt.xlabel("Log N", fontsize = 17)
plt.legend()
plt.ylabel("Log(<d>)", fontsize = 17);
```

Thus, we find that the original data follows the equation $$\text{Distance Travelled} \approx (0.844) N^{0.512}$$

# Part 2: Air Molecule Diffusion in 3D

$\lambda$ $\phi$ $\theta$

```python
def simulation3d(𝜆, time, velocity): #Function to simulate the diffusion of the molecule
    
    steps_taken = 0
    tm = 𝜆 / velocity #The time taken for a collision to take place
    total_steps = int(time / tm) #The total number of collisions that should take place in the given time
    
    history = []
    
    while steps_taken < total_steps:
        
        𝜙 = np.deg2rad(rng.randint(0, 360)) #Generates random direction to move molecule
        𝜃 = np.deg2rad(rng.randint(0, 180))
        
        update = np.array([𝜆 * np.cos(𝜙) * np.sin(𝜃), 𝜆 * np.sin(𝜙) * np.sin(𝜃), 𝜆 * np.cos(𝜃)]) #Updates the position
        history.append(update)
        
        steps_taken += 1
    
    final_position = np.array([0.0, 0.0, 0.0]) #Adds together all the updates to get to a final position
    for i in history:
        final_position += i
    
    distance_travelled = np.linalg.norm(final_position)
    
    return distance_travelled

def mean_distance3d(𝜆, time, velocity, repetitions):
    distance = []
    
    for i in range(repetitions):
        distance.append(simulation3d(𝜆, time, velocity))
        
    mean_distance = np.mean(distance)
    
    return distance, mean_distance


def mean_distance3d𝜆units(𝜆dist, time_taken, velocity, repetitions):
    mean_distance = []
    collisions = []
    
    for i in tqdm(range(1, 𝜆dist + 1)):
        collision_number = int(time_taken / (i / velocity))
        if i > 1:
            if collision_number == int(time_taken / ((i-1) / velocity) ):
                continue
        
        collisions.append(collision_number) 
        mean_distance.append(mean_distance3d(i, time_taken, velocity, repetitions)[1] / i)
        time.sleep(0.1)
        
    return mean_distance, collisions
```

## Varying Mean free path

$$ \frac{<d>}{\lambda} \approx N^{q}$$

If we take the log of both sides, we find that there is a linear relationship between both sides. We shall also take the right side to be a new quanity S

$$ \log{S} = q\log{N}$$

```python
def varymean_linear_regression(logN, q):
    return q * logN
```

```python
varymeanfreepathdata = mean_distance3d𝜆units(500, 1, 500, 200)
```

```python
varylogC = np.log(varymeanfreepathdata[1])
varylogS = np.log(varymeanfreepathdata[0])

varymeanfreepath_params = curve_fit(varymean_linear_regression, varylogC, varylogS)
np.vectorize(varymean_linear_regression)

print("q = " + f"{varymeanfreepath_params[0][0]: .3f}");
```

```python
plt.plot(varylogC, varylogS, color="lightblue", label="Simulation Data")
plt.plot(varylogC, varymean_linear_regression(varylogC, varymeanfreepath_params[0][0]), color="grey", label="Linear Regression")
plt.xlabel("Log N", fontsize = 17)
plt.ylabel("Log(<d>)", fontsize = 17)
plt.legend();
```

# Lil Extension of the 3D model

Here, we simply implement a temperature dependence to the system and thus allow the velocity to change.

The mean speed of a monoatomic, nearly ideal gas is given by:

$$\left\langle v \right\rangle = \sqrt{\frac{8RT}{\pi M}} $$

Where R = 8.31 J/mol K

Let us look at the velocity of diatomic Nitrogen (Primary component of air)

```python
def velocity_dist(T1, T2, M): #Used to generate a distribution of velocities for given molecule between T1 and T2
    velocities = []
    
    for i in range(T1, T2):
        velocities.append(np.sqrt( (8 * 8.31 * i) / (np.pi * M) ) )
    
    return velocities
```

Let's investigate how the distance travelled changes for temperatures from -15C to 32C to see whether there is any significant variation in the range of temperatures seen in a country such as Canada. This is taken in an ideal scenario where the temperature remains constant for the duration of a given path. It is also taken that the mean free path is independent of v

```python
velocities = velocity_dist(258, 305, 0.028)

e_distances = []
for i in velocities:
    e_distances.append(mean_distance3d(1, 1, i, 100))

print(e_distances[0])
```
